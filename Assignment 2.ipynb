```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds = sns.load_dataset("car_crashes")
ds
```

```python
ds.info()
```

```python
ds.head()
```

```python
ds.tail()
```

### Scatter Plot

```python
sns.scatterplot(x='total',y='alcohol',data=ds)
plt.title("Total vs Alocohol")
plt.xlabel("Total Crashes")
plt.ylabel("Consumption of Alocohol")
plt.show()
```

```python
sns.scatterplot(x='total',y='speeding',data=ds)
plt.title("Total vs Speeding")
plt.xlabel("Total Crashes")
plt.ylabel("Speeding")
```

### Line Plot

```python
sns.lineplot(x='total',y='alcohol',data=ds,errorbar=None)
```

```python
sns.lineplot(x='total',y='speeding',data=ds,errorbar=None)
```

### Distribution Plot

```python
sns.distplot(ds['no_previous'])
```

### Box Plot

```python
sns.boxplot(x='total',y='alcohol',data=ds)
```

```python
sns.boxplot(x='total',y='speeding',data=ds)
```

### Relation Plot

```python
sns.relplot(x='total',y='alcohol',data=ds,hue='speeding')
```

```python
sns.relplot(x='total',y='speeding',data=ds,hue='alcohol')
```

### Count Plot

```python
sns.countplot(x='speeding',data=ds)
```

```python
sns.countplot(x='not_distracted',data=ds)
```

### Joint Plot

```python
sns.jointplot(x='total',y='alcohol',data=ds)
```

```python
sns.jointplot(x='total',y='speeding',data=ds)
```

### Bar Plot

```python
sns.barplot(x='speeding',y='alcohol',data=ds)
```
